# `OpenCV` 图像基本操作

## 图像加减

- 加法：两张图片相加使得整体亮度提升，暗处细节更明显，但可能过曝。
- 减法：两张图片相减使得整体亮度降低，也可用于找不同

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

sheep = cv2.imread('../img/sheep.jpg')

if sheep is not None:
    mask = np.ones(sheep.shape, np.uint8) * 50
    mask_2 = np.ones(sheep.shape, np.uint8) * 100
    add_result = cv2.add(sheep, mask)
    add_result_2 = cv2.add(sheep, mask_2)

    plt.figure(figsize=(28, 28))

    plt.subplot(3, 1, 1)
    plt.imshow(sheep)
    plt.title('source')

    plt.subplot(3, 1, 2)
    plt.imshow(add_result)
    plt.title('add_result')

    plt.subplot(3, 1, 3)
    plt.imshow(add_result_2)
    plt.title('add_result_2')

```

```python
zbt_1 = cv2.imread('../img/zbt_1.jpeg')
zbt_2 = cv2.imread('../img/zbt_2.jpeg')

if zbt_1 is not None and zbt_2 is not None:
    diff = cv2.subtract(zbt_1, zbt_2)

    plt.figure(figsize=(18, 18))

    plt.subplot(3, 1, 1)
    plt.imshow(zbt_1)
    plt.title('source 1')

    plt.subplot(3, 1, 2)
    plt.imshow(zbt_2)
    plt.title('source 2')

    plt.subplot(3, 1, 3)
    plt.imshow(diff)
    plt.title('diff')
```
